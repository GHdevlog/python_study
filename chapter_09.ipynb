```python
# 필요한 모듈 임포트
import sqlite3
from tkinter import *
from tkinter import messagebox

# 데이터 삽입 함수 정의
def insertData():
    con, cur = None, None
    data1, data2, data3, data4 = "", "", "", ""  # 데이터 변수 초기화
    sql = ""
    
    # 데이터베이스 연결
    con = sqlite3.connect("C:/Users/user/Downloads/sqlite-tools-win-x64-3450200/naverDB")
    cur = con.cursor()
    
    # 입력된 데이터 가져오기
    data1 = edt1.get()
    data2 = edt2.get()
    data3 = edt3.get()
    data4 = edt4.get()
    
    try:
        # 데이터베이스에 데이터 삽입
        sql = "INSERT INTO userTable VALUES(?, ?, ?, ?)"
        cur.execute(sql, (data1, data2, data3, data4))
    except:
        messagebox.showerror("오류", "데이터 입력 오류 발생")
    else:
        messagebox.showinfo("성공", "데이터 입력 성공")
    
    # 변경사항 저장 및 연결 종료
    con.commit()
    con.close()

# 데이터 조회 함수 정의
def selectData():
    strData1, strData2, strData3, strData4 = [],[],[],[]
    con = sqlite3.connect("C:/Users/user/Downloads/sqlite-tools-win-x64-3450200/naverDB")
    cur = con.cursor()
    cur.execute("SELECT * FROM userTable")
    
    # 헤더 추가
    strData1.append("사용자 ID")
    strData2.append("사용자 이름")
    strData3.append("이메일")
    strData4.append("출생연도")
    
    strData1.append("---------")
    strData2.append("---------")
    strData3.append("---------")
    strData4.append("---------")
    
    # 데이터 읽어오기
    while True:
        row = cur.fetchone()
        if row == None: break
        strData1.append(row[0])
        strData2.append(row[1])
        strData3.append(row[2])
        strData4.append(row[3])
        
    # 리스트박스 초기화
    listData1.delete(0, listData1.size()-1)
    listData2.delete(0, listData2.size()-1)
    listData3.delete(0, listData3.size()-1)
    listData4.delete(0, listData4.size()-1)
    
    # 리스트박스에 데이터 삽입
    for item1, item2, item3, item4 in zip(strData1, strData2, strData3, strData4):
        listData1.insert(END, item1)
        listData2.insert(END, item2)
        listData3.insert(END, item3)
        listData4.insert(END, item4)
    
    # 데이터베이스 연결 종료
    con.close()

# 메인 윈도우 생성
window = Tk()
window.geometry("600x300")
window.title("GUI 데이터 입력")

# 입력 프레임 생성
edtFrame = Frame(window)
edtFrame.pack()

# 리스트 프레임 생성
listFrame = Frame(window)
listFrame.pack(side=BOTTOM, fill=BOTH, expand=1)

# 입력 필드 생성
edt1 = Entry(edtFrame, width=10); edt1.pack(side=LEFT, padx=10, pady=10)
edt2 = Entry(edtFrame, width=10); edt2.pack(side=LEFT, padx=10, pady=10)
edt3 = Entry(edtFrame, width=10); edt3.pack(side=LEFT, padx=10, pady=10)
edt4 = Entry(edtFrame, width=10); edt4.pack(side=LEFT, padx=10, pady=10)

# 입력 및 조회 버튼 생성
btnInsert = Button(edtFrame, text="입력", command=insertData)
btnInsert.pack(side=LEFT, padx=10, pady=10)
btnSelect = Button(edtFrame, text="조회", command=selectData)
btnSelect.pack(side=LEFT, padx=10, pady=10)

# 리스트박스 생성
listData1 = Listbox(listFrame, bg='red')
listData1.pack(side=LEFT, fill=BOTH, expand=1)
listData2 = Listbox(listFrame, bg='green')
listData2.pack(side=LEFT, fill=BOTH, expand=1)
listData3 = Listbox(listFrame, bg='blue')
listData3.pack(side=LEFT, fill=BOTH, expand=1)
listData4 = Listbox(listFrame, bg='yellow')
listData4.pack(side=LEFT, fill=BOTH, expand=1)

# 이벤트 루프 실행
window.mainloop()
```

```python
import sqlite3

# SQLite 데이터베이스에 연결합니다.
conn = sqlite3.connect("C:/Users/user/Downloads/sqlite-tools-win-x64-3450200/naverDB")

# 커서를 생성합니다.
cursor = conn.cursor()

# 컬럼명을 조회하고자 하는 테이블의 이름을 지정합니다.
table_name = 'userTable'

# PRAGMA 문을 실행하여 컬럼 정보를 가져옵니다.
cursor.execute(f"PRAGMA table_info({table_name})")

# 결과를 모두 가져옵니다.
columns = cursor.fetchall()

# 컬럼 정보를 출력합니다.
for column in columns:
    print(column)  # 컬럼명은 두 번째 인덱스에 있습니다.

# 연결을 닫습니다.
conn.close()
```
