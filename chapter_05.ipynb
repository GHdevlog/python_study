```python
from tkinter import *
from tkinter.filedialog import *

def func_open():
    filename = askopenfilename(parent=window, filetypes=(("GIF 파일","*.gif"),("모든 파일","*.*")))
    photo = PhotoImage(file=filename)

    width = photo.width()
    height = photo.height()

    for x in range(width):
        for y in range(height):
            r,g,b = photo.get(x,y)
            gray = int((r+g+b)/3)
            photo.put("#%02x%02x%02x" % (gray,gray,gray), (x,y))

    pLable.configure(image=photo)
    pLable.image = photo

def func_exit():
    window.quit()
    window.destroy()

window = Tk()
window.geometry("500x500")
window.title("명화 감상")

photo = PhotoImage()
pLable = Label(window, image=photo)
pLable.pack(expand=1, anchor=CENTER)

mainMenu = Menu(window)
window.config(menu=mainMenu)

fileMenu = Menu(mainMenu)
mainMenu.add_cascade(label="파일", menu=fileMenu)

fileMenu.add_command(label="파일 열기",command=func_open)
fileMenu.add_separator()
fileMenu.add_command(label="프로그램 종료",command=func_exit)

window.mainloop()
```

```python
from tkinter import *

def loadImage(fname):
    global inImage, XSIZE, YSIZE
    fp = open(fname, 'rb')
    
    for i in range(XSIZE):
        tmpList = []
        for j in range(YSIZE):
            data = int(ord(fp.read(1)))
            tmpList.append(data)
        inImage.append(tmpList)
    
    fp.close()

def displayImage(image):
    global XSIZE, YSIZE
    rgbString = ""
    for i in range(XSIZE):
        tmpString = ""
        for j in range(YSIZE):
            data = image[i][j]
            tmpString += "#%02x%02x%02x" % (data,data,data)
        rgbString += "{"+ tmpString + "}"
        paper.put(rgbString)

window = None
canvas = None
XSIZE, YSIZE = 256,256
inImage = []

window = Tk()
window.title("흑백 사진 보기")
canvas = Canvas(window, height=XSIZE, width=YSIZE)
paper = PhotoImage(width=XSIZE, height=YSIZE)
canvas.create_image((XSIZE/2, YSIZE/2), image=paper, state="normal")

filename = "../source/RAW/dog256.raw"
loadImage(filename)

displayImage(inImage)

canvas.pack()
window.mainloop()
```

```python
import tkinter as tk
from tkinter import filedialog
from PIL import Image, ImageTk

def open_and_convert_image():
    global label_image  # label_image 변수를 전역 변수로 선언
    # 파일 대화 상자를 열어 이미지 파일 선택
    file_path = filedialog.askopenfilename()
    if file_path:
        # 이미지 열기
        image = Image.open(file_path)
        # 이미지를 흑백으로 변환
        bw_image = image.convert("L")
        # 이미지 크기 조정
        bw_image.thumbnail((300, 300))
        # Tkinter에서 사용할 수 있도록 이미지 변환
        photo = ImageTk.PhotoImage(bw_image)
        # 이미지를 보여줄 라벨 생성 또는 업데이트
        if label_image is None:
            label_image = tk.Label(root, image=photo)
            label_image.photo = photo  # 이미지가 가비지 컬렉션되는 것을 방지
            label_image.pack()
        else:
            label_image.configure(image=photo)
            label_image.photo = photo

# Tkinter 윈도우 생성
root = tk.Tk()
root.title("이미지 흑백 변환 예제")

# 메뉴 바 생성
menubar = tk.Menu(root)
root.config(menu=menubar)

# 파일 메뉴 생성
file_menu = tk.Menu(menubar, tearoff=False)
menubar.add_cascade(label="파일", menu=file_menu)

# 파일 메뉴에 파일 열기 옵션 추가
file_menu.add_command(label="열기", command=open_and_convert_image)

# 이미지를 표시할 라벨 변수
label_image = None

# 윈도우 실행
root.mainloop()
```

### GPT와 구현해보기

```python
import tkinter as tk
from tkinter import filedialog

file_path = None  # 열었던 파일의 경로를 저장할 전역 변수

def open_file():
    global file_path
    file_path = filedialog.askopenfilename()
    if file_path:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
        text.delete('1.0', tk.END)  # 기존 텍스트 삭제
        text.insert(tk.END, content)  # 파일 내용 표시

def save_file():
    global file_path
    if file_path:
        content = text.get('1.0', tk.END)
        with open(file_path, 'w', encoding='utf-8') as file:
            file.write(content)
    else:
        save_as_file()

def save_as_file():
    global file_path
    file_path = filedialog.asksaveasfilename(
        defaultextension=".txt",
        filetypes=[("텍스트 파일", "*.txt"), ("모든 파일", "*.*")]  # 사용자가 선택할 수 있는 파일 형식 목록
    )
    if file_path:
        content = text.get('1.0', tk.END)
        with open(file_path, 'w', encoding='utf-8') as file:
            file.write(content)

root = tk.Tk()
root.title("파일 메뉴 예제")

# 메뉴바 생성
menubar = tk.Menu(root)
root.config(menu=menubar)

# 파일 메뉴 생성
file_menu = tk.Menu(menubar, tearoff=False)
menubar.add_cascade(label="파일", menu=file_menu)

# 파일 메뉴에 파일 열기 기능 추가
file_menu.add_command(label="열기", command=open_file)

# 파일 메뉴에 구분선 추가
file_menu.add_separator()

# 파일 메뉴에 파일 저장 기능 추가
file_menu.add_command(label="저장", command=save_file)

# 파일 메뉴에 다른 이름으로 저장 기능 추가
file_menu.add_command(label="다른 이름으로 저장", command=save_as_file)

# 파일 메뉴에 구분선 추가
file_menu.add_separator()

# 파일 메뉴에 종료 메뉴 추가
file_menu.add_command(label="종료", command=root.destroy)

# 텍스트 위젯 생성
text = tk.Text(root)
text.pack(fill=tk.BOTH, expand=True)

root.mainloop()
```
