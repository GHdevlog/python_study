```python
import timeit
import random

SIZE = 5
pythonList = [[random.randint(0,255) for _ in range(SIZE)] for _ in range(SIZE)]

for i in range(SIZE):
    for j in range(SIZE):
        print("%3d" % pythonList[i][j], end=' ')
    print()
print()

for i in range(SIZE):
    for j in range(SIZE):
        pythonList[i][j] += 100
        
code = '''
import random

SIZE = 5
pythonList = [[random.randint(0,255) for _ in range(SIZE)] for _ in range(SIZE)]
for i in range(SIZE):
    for j in range(SIZE):
        pythonList[i][j] += 100'''
        
myTimer = timeit.Timer(code)
execution_time = myTimer.repeat(repeat=5, number=10)

for time in execution_time:
    print("%e" % time)
print()

for i in range(SIZE):
    for j in range(SIZE):
        print("%3d" % pythonList[i][j], end=' ')
    print()
```

```python
import numpy as np
import random

# 0부터 255 사이의 랜덤 정수로 이루어진 5개의 요소를 가진 파이썬 리스트 생성
pythonList = [random.randint(0,255) for _ in range(5)]
print("* 파이썬 리스트 -->", pythonList)

# 파이썬 리스트를 NumPy 배열로 변환
numpyAry1 = np.array(pythonList)
print("* arry(pythonList) --> ", numpyAry1)

# 0부터 4까지의 정수로 이루어진 NumPy 배열 생성
numpyAry2 = np.arange(5)
print("* arange(5) -->",numpyAry2)

# 3부터 7까지의 정수로 이루어진 NumPy 배열 생성
numpyAry3 = np.arange(3,8)
print("* np.arange(3,8)", numpyAry3)

# 0부터 99까지 20씩 증가하는 정수로 이루어진 NumPy 배열 생성
numpyAry3 = np.arange(0,100,20)
print("* arange(0,100,20) -->", numpyAry3)

# 모든 요소가 1인 1차원 NumPy 배열 생성
numpyAry4 = np.ones(5)
print("* ones(5)-->",numpyAry4)

# 모든 요소가 1인 3x4 크기의 2차원 NumPy 배열 생성
numpyAry5 = np.ones((3,4))
print("* ones((3,4))-->", numpyAry5)

# 모든 요소가 0인 1차원 NumPy 배열 생성
numpyAry6 = np.zeros(5)
print("* zeros(5)-->",numpyAry6)

# 초기화되지 않은 값으로 채워진 1차원 NumPy 배열 생성
numpyAry7 = np.empty(6)
print("* empty(6)-->",numpyAry7)

# 모든 요소가 지정한 값(여기서는 33)으로 채워진 1차원 NumPy 배열 생성
numpyAry8 = np.full(5,33)
print("* full(5,33)-->",numpyAry8)

# 대각선 요소가 1이고 나머지 요소가 0인 5x5 크기의 2차원 NumPy 배열 생성
numpyAry9 = np.identity(5)
print("* identity(5)-->",numpyAry9)
```

```python
import random

# 기본 리스트 크기 및 시작 행/열 설정
SIZE = 5
startRow, startCol = 1, 1
nSIZE = 3

# 리스트 초기화 및 값 채우기
value = 1
myList1 = []
for _ in range(SIZE):
    tmpList = []
    for _ in range(SIZE):
        tmpList.append(value)
        value += 1
    myList1.append(tmpList)

# myList1 출력
for i in range(SIZE):
    [print("%3d" % myList1[i][j], end=' ') for j in range(SIZE)]
    print()
print()

# myList1에서 일부 영역을 가져와 myList2에 저장
myList2 = []
for i in range(startRow, startRow + nSIZE):
    tmpList = []
    for j in range(startCol, startCol + nSIZE):
        tmpList.append(myList1[i][j])
    myList2.append(tmpList)

# myList2 출력
for i in range(nSIZE):
    [print("%3d" % myList2[i][j], end=' ') for j in range(nSIZE)]
    print()
print()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


inFilename = "source/Excel/singer.xlsx"
outFilename = "source/Excel/singer_over6.xlsx"

df_senior = pd.read_excel(inFilename, 'senior', index_col=None)
df_junior = pd.read_excel(inFilename, 'junior', index_col=None)

df_singer =pd.concat([df_senior,df_junior])
df_singer_over6 = df_singer[df_singer['인원']>=6]
df_singer_over6 = df_singer_over6.sort_values(by=['인원'], axis=0, ascending=False)

df_singer_over6 = df_singer_over6.loc[:,['아이디','이름','인원','유튜브 조회수']]

x_data = df_singer_over6['아이디']
y_data = df_singer_over6['인원']
colorAry = [np.random.choice(['red','green','blue','brown','gold','lime','aqua','magenta','purple']) for _ in range(len(x_data))]

plt.bar(x_data,y_data,color=colorAry)
plt.show()

writer = pd.ExcelWriter(outFilename)
df_singer_over6.to_excel(writer, sheet_name='singer', index=False)
writer.close()
print("Save. OK~")
```
